# Opioid Death and Employment Regressions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from scipy.stats import pearsonr
```

```python
death_unemployment = pd.read_csv('../data/Merged_complete_employment_death.csv', index_col=[0])
```

```python
no_2020 = death_unemployment[death_unemployment["Year"]!= 2020]
```

```python
all_means= no_2020.groupby(["County"]).mean()
```

```python
opioid_death= death_unemployment["Opioid Death"]
unemployment = death_unemployment["Unemployment"]


f, ax = plt.subplots(figsize =(9, 8))
plt.plot(unemployment,opioid_death, 'o')
m, b = np.polyfit(unemployment,opioid_death, 1)
plt.plot(unemployment, m*unemployment + b)

plt.xlabel("Unemployment", size=15)
plt.ylabel("Opioid Death", size=15)
plt.title("Opioid Death as Unemployment grows", size=18);
```
